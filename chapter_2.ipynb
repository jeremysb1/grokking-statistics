```python
# Three exams and a final score
exam_scores = [85, 90, 88, 92]
```

```python
# Weighting for each exam
weights = [.2, .2, .2, .4]
```

```python
# Calculate weighted average
weighted_average = sum(e*w for e, w in zip(exam_scores, weights)) / sum(weights)
```

```python
print("Final grade: ", weighted_average)
```

```python
print("Using NumPy: ", np.average(exam_scores, weights=weights))
```

```python
temps = [77,75,68,71,60,65,81,71,56,73,
       79,80,85,84,72,69,82,81,71,63,
       75,83,80,80,88,76,79,73,80,81,82]
```

```python
mean = np.mean(temps)
```

```python
print("MIN: ", np.min(temps), "MAX: ", np.max(temps))
```

```python
print("MEAN: ", mean)
```

```python
print("MEDIAN: ", np.median(temps))
```

```python
# sum all the squared differences
squared_diffs = 0
for t in temps:
    squared_diffs += (t - mean) ** 2
```

```python
# take the avg of the squared differences
variance = squared_diffs / (len(temps) - 1)

print("Variance: ", variance)
```

```python
from math import sqrt

def variance(data: [int]):
  mean = sum(data) / len(data)

  squared_diffs = 0
  for x in data:
      squared_diffs += (x - mean) ** 2

  return squared_diffs / (len(data) - 1)

def standard_deviation(data: [int]):
  return sqrt(variance(data))

print("Standard Deviation: ", standard_deviation(temps))
```
